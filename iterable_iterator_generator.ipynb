# Generator
* we are trying to generate something
* less memory occupied; less time to execute
* but in function huge memory occupied
* yield create a generator; complete memory management with less time 

```python
list(range(0,45,3))
```

```python
def gencube(n):
    l=[]
    for i in range(n):
        #print(i**3)
        l.append(i**3)
    return l
```

```python
#it takes huge amount of time
gencube(6)
```

```python
#using yield, not range()
# yield create a generator
def gencube(n):
    for i in range(n):
        yield i**3
```

```python
gencube(6)
```

```python
for i in gencube(9000):
    print(i)
```

Task: generate the next number which is sum previuos two number
Fibonacci sequence

```python
def fib_num(n):
    a=1
    b=1
    for i in range(n):
        yield a,i
        a,b=b,a+b
```

```python
for i in fib_num(1000):
    print(i)
```

```python
def fib1(n):
    a=1
    b=1
    output=[]
    for i in range(n):
        output.append(a)
        a,b=b,a+b
    return output
```
